```python
import pandas as pd

# Load your datasets
socioeconomic = pd.read_csv("socio_economic.csv", dtype=str)
health = pd.read_csv("500_Cities__Census_Tract-level_Data__GIS_Friendly_Format___2019_release_20250317.csv", dtype=str)
spatial = gpd.read_file('census_with_healthcare_counts.geojson')

# Ensure the geoid columns are of the same type (e.g., string)
socioeconomic['geoid'] = socioeconomic['GEOID'].astype(str)
health['geoid'] = health['TractFIPS'].astype(str)
spatial['geoid'] = spatial['GEOID'].astype(str)

# Get the set of geoids for each dataset
geoids_socio = set(socioeconomic['geoid'])
geoids_health = set(health['geoid'])
geoids_spatial = set(spatial['geoid'])

# Find the common geoids across all three datasets
common_geoids = geoids_socio.intersection(geoids_health).intersection(geoids_spatial)

# Optionally, filter each dataset to only include these common geoids
socioeconomic_common = socioeconomic[socioeconomic['geoid'].isin(common_geoids)]
health_common = health[health['geoid'].isin(common_geoids)]
spatial_common = spatial[spatial['geoid'].isin(common_geoids)]
```

```python
import pandas as pd

# Load your socioeconomic dataset
df = socioeconomic_common

# List of columns that need to be converted to numeric.
# (Update the list if you have additional columns to convert.)
numeric_columns = ['Total_Population', 'Total_Households',
    'Median_Household_Income', 'Poverty_Population', 'Unemployed_Population',
    'Per_Capita_Income', 'SNAP_Recipients', 'Total_Pop_Health_Insurance',
    'Uninsured_Population', 'Medicaid_Coverage', 'Medicare_Coverage',
    'Private_Insurance_Coverage', 'High_School_Graduates', 'Bachelor_Degree_Holders',
    'Less_than_High_School', 'Households_No_Vehicle', 'Commute_Time',
    'Public_Transit_Usage', 'Renter_Occupied_Housing', 'Rent_as_Income_Percentage',
    'Overcrowded_Housing', 'Elderly_Population', 'Children_Population',
    'White_Population', 'Black_Population',
    'Limited_English_Proficiency', 'Households_No_Internet'
]

# Function to clean and convert a column
def clean_and_convert(column):
    # Remove commas and dollar signs
    col_clean = df[column].str.replace('[\$,]', '', regex=True)
    # Optionally, remove percentage signs if present
    col_clean = col_clean.str.replace('%', '', regex=False)
    # Convert the cleaned string to numeric, coercing errors into NaN
    df[column] = pd.to_numeric(col_clean, errors='coerce')

# Apply the cleaning function to each numeric column
for col in numeric_columns:
    clean_and_convert(col)

# Verify the conversions by checking data types and summary statistics
print(df[numeric_columns].dtypes)
print(df[numeric_columns].describe())
```

```python
import pandas as pd
import numpy as np

# Assume df is your original DataFrame loaded with the socioeconomic data
# For example:
# df = pd.read_csv('socioeconomic_data.csv')

# --- 1. Create Transformed/Derived Features ---

# 1a. Log transform the Median Household Income to reduce right skew
# (Using np.log1p to handle any zero values safely)

# 1b. Compute Uninsured Rate as a percentage
# Assuming that Uninsured_Population is the count of uninsured and 
# Total_Pop_Health_Insurance represents the count of insured individuals.
# The total population related to insurance becomes: total = insured + uninsured.
df['Uninsured_Rate'] = (df['Uninsured_Population'] / 
                        (df['Total_Pop_Health_Insurance'] + df['Uninsured_Population'])) * 100

# 1c. (Optional) If you have a total households column (e.g., 'Total_Households'),
# you can convert Households_No_Vehicle and Households_No_Internet into percentages.
# Uncomment and adjust the following if you have that column:
df['No_Vehicle_Rate'] = (df['Households_No_Vehicle'] / df['Total_Households']) * 100
df['No_Internet_Rate'] = (df['Households_No_Internet'] / df['Total_Households']) * 100
df['Limited_English_Proficiency_Rate'] = (df['Limited_English_Proficiency'] / df['Total_Population']) * 100

# If no denominators are available, you might keep raw counts for now, but rates are preferred for comparability.

# --- 2. Select Recommended Features ---
# Based on our feature selection recommendations, we want to keep the following columns:
#  - Geography, state, county, tract, GEOID (for spatial and identifier purposes)
#  - Log_Median_Household_Income (as the income proxy)
#  - Uninsured_Rate (as the indicator for insurance coverage gaps)
#  - Limited_English_Proficiency (if available as count; ideally you convert to a rate if total population is known)
#  - Households_No_Vehicle (or No_Vehicle_Rate if calculated)
#  - Households_No_Internet (or No_Internet_Rate if calculated)
#  - Disability_Population (or Disability_Rate if calculated)
#  - Rent_as_Income_Percentage (as a proxy for housing affordability)

# Adjust the selection based on whether you computed rates or not. Hereâ€™s one way:
columns_to_keep = [
    'Geography', 'state', 'county', 'tract', 'GEOID', 'Total_Households', 'Total_Population',
    'Median_Household_Income',
    'Uninsured_Rate',
    'Limited_English_Proficiency_Rate',  # If you compute a rate, replace with that column (e.g., Limited_English_Rate)
    'No_Vehicle_Rate',        # or 'No_Vehicle_Rate' if computed
    'No_Internet_Rate',       # or 'Disability_Rate' if computed
    'Rent_as_Income_Percentage'
]

final_df = df[columns_to_keep].copy()

# --- 3. Save the Final Dataset ---
# You can now save the final dataset for modeling.
final_df.to_csv('final_socio_economic.csv', index=False)

print("Final dataset created with the following columns:")
print(final_df.columns)
```

```python
import geopandas as gpd
import pandas as pd

# Load your socioeconomic dataset (the one you created earlier)
df_final = pd.read_csv("final_socio_economic.csv",  dtype=str)  # or use your current dataset

# Load the GeoJSON file containing tract centroids
# Make sure the filename matches your actual file
centroids = gpd.read_file("census_tract_centroids.geojson",  dtype=str)

# Extract latitude and longitude from the geometry column.
# In GeoJSON, coordinates are in (longitude, latitude) order.
centroids["longitude"] = centroids.geometry.x
centroids["latitude"] = centroids.geometry.y

# Select only the necessary columns for merging: GEOID, latitude, longitude.
centroids = centroids[["GEOID", "latitude", "longitude"]]

# Merge the centroids with your socioeconomic data on GEOID.
# If your socioeconomic data already has some lat/lon columns, they will be overwritten or suffixed.
df_final = df_final.merge(centroids, on="GEOID", how="left")

# Check the merge result
print(df_final.head())

# Save the merged dataset if desired
df_final.to_csv("socio_economic_with_latlon.csv", index=False)
print("Merged data saved to 'socio_economic_with_latlon.csv'")
```

```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
from folium import IFrame

# Load the dataset with latitude and longitude
df = pd.read_csv("socio_economic_final.csv")

# Determine a central location for the map
map_center = [df['latitude'].mean(), df['longitude'].mean()]

# Create a Folium map with a modern tile (e.g., CartoDB Positron)
m = folium.Map(location=map_center, zoom_start=5, tiles='cartodbpositron')

# Initialize a MarkerCluster
marker_cluster = MarkerCluster().add_to(m)

# Loop through the DataFrame and add markers with a modern styled popup
for idx, row in df.iterrows():
    # Create a modern popup using HTML/CSS
    html = f"""
    <html>
      <head>
        <style>
          .popup-container {{
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            background: #ffffff;
            padding: 10px;
            border-radius: 8px;
            box-shadow: 0 2px 6px rgba(0, 0, 0, 0.15);
            width: 260px;
          }}
          .popup-header {{
            font-size: 16px;
            font-weight: bold;
            margin-bottom: 8px;
            color: #2c3e50;
          }}
          .popup-content {{
            font-size: 14px;
            color: #34495e;
            line-height: 1.4;
          }}
          .popup-content p {{
            margin: 4px 0;
          }}
          .popup-content b {{
            color: #2c3e50;
          }}
        </style>
      </head>
      <body>
        <div class="popup-container">
          <div class="popup-header">{row['Geography']}</div>
          <div class="popup-content">
            <p><b>State:</b> {row['state']}</p>
            <p><b>County:</b> {row['county']}</p>
            <p><b>Median Income:</b> ${row['Median_Household_Income']}</p>
            <p><b>Uninsured Rate:</b> {row['Uninsured_Rate']}%</p>
            <p><b>English Proficiency:</b> {row['Limited_English_Proficiency_Rate']}</p>
            <p><b>No Vehicle Rate:</b> {row['No_Vehicle_Rate']}%</p>
            <p><b>No Internet Rate:</b> {row['No_Internet_Rate']}%</p>
            <p><b>Rent as Income %:</b> {row['Rent_as_Income_Percentage']}%</p>
          </div>
        </div>
      </body>
    </html>
    """
    iframe = IFrame(html, width=280, height=380)
    popup = folium.Popup(iframe, max_width=300)
    
    # Add a marker with a custom icon and tooltip
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=popup,
        tooltip=row['Geography'],
        icon=folium.Icon(icon="info-sign", color="blue")
    ).add_to(marker_cluster)

# Save the interactive map to an HTML file
m.save("modern_interactive_map.html")
print("Interactive map saved as 'modern_interactive_map.html'")
```
