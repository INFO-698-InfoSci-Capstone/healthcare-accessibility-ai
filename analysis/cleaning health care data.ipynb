```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
socioecon = pd.read_csv(
    "./data/rawData/socio_economic.csv"
)
```

```python
df1 = socioecon
df2 = t

print(
    len(df1[df1['GEOID'].isin(df2['TractFIPS'])]),
    len(df2[df2['TractFIPS'].isin(df1['GEOID'])])
)
```

```python
t_subset = df2[df2['TractFIPS'].isin(df1['GEOID'])]
```

```python
t_subset
```

```python
cols2remove = []
exempt_cols = [
    "StateAbbr", "PlaceName",
]

for i in t_subset.columns:
    if type(t_subset[i][0]) == str and i not in exempt_cols:
        cols2remove.append(i)
```

```python
cols2remove
```

```python
t_subset = t_subset.drop(columns=cols2remove)
```

```python
tforcorr = t_subset.drop(columns=exempt_cols)  
```

```python
plt.figure(figsize=(25, 20))
sns.heatmap(tforcorr.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Independent Variables")
plt.show()
```
